### Task - 3 (REMOVAL OF STOP WORDS)

```python
result = []
stop_words = set(stopwords.words('english'))
```

```python
def removeStopWords(para):
  words = nltk.word_tokenize(para)
  for i in words:
    if i not in stop_words:
      result.append(i)
      
```

```python
for  para in dataframe['text']:
        removeStopWords(str(para))
result
```
